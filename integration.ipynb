```python
from scipy.integrate import quad
from scipy.special import jv
import numpy as np
import matplotlib.pyplot as plt
def integrand(xi, x, y):
    return np.log( (x - xi)**2 + y**2 )


a = 0
b = 2
x = 1
y = 1e-5
I = quad(integrand, a, b, args=(x,y))
print(I)
analytical = -(x-b) * np.log( (x-b)**2 + y**2 ) + (x-a) * np.log( (x-a)**2 + y**2 ) - 2*y * np.arctan( (x-b)/y) \
            + 2*y * np.arctan( (x-a)/y) + 2*(x-b) - 2*(x-a)
print(analytical)
```

```python
for k in range(2, 15):
    N = 2**k + 1
    xi = np.linspace( 0, 2, N ) 
    dxi = xi[2] - xi[0]
    f = integrand( xi, x, y )
    I = 0.0
    for i in range( 1, N, 2 ):
        I = I + ( f[i-1] + 4*f[i] + f[i+1] ) * dxi / 6
    print( "N = " + str(N), "I = " + str(I) )
```

```python
v = 0
z = np.linspace(0, 100, 1000)
j = jv(v, z) 

fig, ax = plt.subplots(figsize=(12, 4))
plt.grid(b=True, which='major', linestyle='--', linewidth=.5)
plt.grid(b=True, which='minor', linestyle='--', linewidth=.1)
ax.plot( z, j )
```
